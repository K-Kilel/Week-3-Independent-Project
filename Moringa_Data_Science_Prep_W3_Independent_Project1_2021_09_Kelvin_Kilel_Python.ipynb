Loading the datasets

```python
#Loading Cells Geo csv file

df1 = pd.read_csv('cells_geo.csv', delimiter= ';')

df1.head(5)
```

Loading the Three Telecom Datasets

```python
# Loading the telecom dataset description 

df2 = pd.read_excel('CDR_description.xlsx', sheet_name= 0)
df2
```

```python
# Loading the first dataset 

df4 = pd.read_csv('Telcom_dataset.csv')
df4.head(3)
```

```python
# Loading the second dataset 
df5 = pd.read_csv('Telcom_dataset2.csv')
df5.head(5)
```

```python
# Loading the third dataset 
df6 = pd.read_csv('Telcom_dataset3.csv')
df6.head(5)
```

Cleaning the Datasets
First, we have to ensure the columns for the three datasets are similar to make it easier to merge them.

```python
# From the first dataset let's rename the Product column and the date_time column

df4 = df4.rename(columns={'PRODUTC':'PRODUCT', 'DATETIME': 'DATE_TIME'})
df4.head()
```

```python
# From the second dataset let's rename the DW_A_Number column and the DW_B_Number  column
df5 = df5.rename(columns={'DW_A_NUMBER':'DW_A_NUMBER_INT', 'DW_B_NUMBER': 'DW_B_NUMBER_INT'})
df5.head()
```

```python
# From the third dataset let's rename the cellid column and the siet_id column
df6 = df6.rename(columns={'CELLID':'CELL_ID', 'SIET_ID': 'SITE_ID'})
df6.head()
```

Merging the Datasets

```python
# Merging the three telecom datasets
telecom_merge = pd.concat([df4,df5,df6], ignore_index= True)
telecom_merge
```

Ensuring the merged dataset has the required data type format

```python
# Converting Nan to 0
values = {'VALUE': 0}
telecom_merge = telecom_merge.fillna(value = values)
```

```python
# Converting Nan to 0
values = {'CELL_ON_SITE': 0}
telecom_merge = telecom_merge.fillna(value = values)
```

```python
# Converting the data to the required format.

telecom_merge = telecom_merge.astype({'PRODUCT': str, 'VALUE': int, 'DATE_TIME': str, 'CELL_ON_SITE': str, 'DW_A_NUMBER_INT': str, 'DW_B_NUMBER_INT': str, 'COUNTRY_A': str, 'COUNTRY_B': str, 'CELL_ID': str,'SITE_ID': str})
telecom_merge.head()
```

Merging Telecom merged dataset with the cell geo dataset 

```python
tel_merge = telecom_merge.rename(columns={'SITE_ID':'SITE_CODE'})
tel_merge.head()
```

```python
merged_data = pd.merge(left=tel_merge, right=df1, how='left', left_on='SITE_CODE', right_on='SITE_CODE')
merged_data
```

Analysis of the data

The city that used the services the most for the three days?

```python
merged_data.groupby('VILLES')['VILLES'].count().sort_values(ascending = False)
```

The most used product to the least used product

```python
merged_data.groupby(['PRODUCT'])['VALUE'].count().sort_values(ascending = False)
```

Which cities used which product mostly? 

```python
merged_data.groupby(['PRODUCT', 'VILLES'])['PRODUCT'].count().sort_values(ascending = False)
```
